```python
import csv
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
import nltk
from nltk.corpus import wordnet as wn
from nltk.stem.wordnet import WordNetLemmatizer
```

```python
final_list = []
new_list = ''
with open("C:/Users/Namrata/OneDrive/Desktop/Dataset.txt", 'r') as doc: 
    for lines in doc.readlines(): 
        #print(lines)
        text = nlp(lines)
        for token in text:
            if token.pos_ == 'VERB':
                new_text = WordNetLemmatizer().lemmatize(token.text, 'v')
                new_list = lines.replace(token.text , new_text)
                new_list = ' '.join([lines, new_list])
            else:
                if new_list:
                    final_list.append(new_list)
                    new_list = ''
                
final_list            
                

                
 
                    
```
