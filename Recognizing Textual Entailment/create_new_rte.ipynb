```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/drive/MyDrive/validation.csv')
df = df.iloc[:200]
df = df.drop(columns=['id','article'])
```

```python
df = df.rename(columns={"highlights": "premise"})
df["hypothesis"] = np.nan
df['idx'] = df.index
one = np.full(shape=100, fill_value=1, dtype=np.int)
zero = np.full(shape=100, fill_value=0, dtype=np.int)
label = np.concatenate((one, zero), axis=None)
df["label"] = label.tolist()
df
```

```python
new_premise=[]
for i in df['premise']:
  new = i.replace('\n', '')
  new_premise.append(new)

df['premise'] = new_premise
df.iloc[0,0]
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/drive/MyDrive/snli_1.0_test.csv')
df = df[['gold_label', 'sentence1', 'sentence2']]
```

```python
df = df.rename(columns={"sentence1": "premise", "sentence2": "hypothesis"})
df
```

```python
entailment = df.loc[df['gold_label'] == 'entailment'][:100].reset_index(drop=True)
entailment
```

```python
entailment['label'] = np.full(shape=100, fill_value=0, dtype=np.int)
entailment = entailment.drop(columns=['gold_label'])
entailment
```

```python
not_entailment = df.loc[df['gold_label'] == 'contradiction'][150:250].reset_index(drop=True)
not_entailment['label'] = np.full(shape=100, fill_value=1, dtype=np.int)
not_entailment = not_entailment.drop(columns=['gold_label'])
not_entailment
```

```python
df = pd.concat([entailment, not_entailment])
df['idx'] = df.index
df
```

```python
from sklearn.utils import shuffle
df = shuffle(df)
df = df[['hypothesis', 'idx', 'label', 'premise']]
# df = df.drop(columns=['idx'])
df
```
